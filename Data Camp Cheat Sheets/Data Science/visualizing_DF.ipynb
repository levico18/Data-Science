```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
avocados = pd.read_pickle("D:/Data Science/Data Sources/Data Camp/avoplotto.pkl")

# Look at the first few rows of data
print(avocados.head())

print(avocados.info())

# Get the total number of avocados sold of each size
nb_sold_by_size = avocados.groupby('size')['nb_sold'].sum() # groupby size summing sold avocados

# Create a bar plot of the number of avocados sold by size
nb_sold_by_size.plot(kind = 'bar')

plt.show()
```

```python
# Get the total number of avocados sold on each date
nb_sold_by_date = avocados.groupby('date')['nb_sold'].sum()

# Create a line plot of the number of avocados sold by date
nb_sold_by_date.plot(kind = 'line', x = 'date', y = 'nb_sold')

# Show the plot
plt.show()
```

```python
# Scatter plot of avg_price vs. nb_sold with title
avocados.plot(kind = 'scatter', x = 'nb_sold', y = 'avg_price', title = "Number of avocados sold vs. average price")

# Show the plot
plt.show()
```

```python
# Histogram of conventional avg_price 
avocados[avocados['type'] == 'conventional']['avg_price'].hist(alpha = 0.5, bins = 20) # subset to avg_price for all conventional type

# Histogram of organic avg_price
avocados[avocados['type'] == 'organic']['avg_price'].hist(alpha = 0.5, bins = 20) # subset to avg_price for all organic type

# Add a legend
plt.legend(['conventional','organic'])

# Show the plot
plt.show()
```
