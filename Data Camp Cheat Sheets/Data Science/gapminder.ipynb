```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\Data Science\Data Sources\Data Sets\Gapminder.csv")

print(df.year)
```

```python
plt.plot(df.gdp_cap,df.life_exp)
plt.show() #plot does not make sense
```

```python
#relationship between capita and life exp


plt.scatter(df.gdp_cap,df.life_exp, alpha= 0.8)

#tick labels for the x-asxi
x_tick_value = [1000, 10000, 100000]
x_tick_lab = ['1k', '10k', '100k']

plt.xscale('log') #put logarithmic value for x axis
plt.xlabel('GDP per Cap [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks(x_tick_value,x_tick_lab) #applying tick labels to x axis
plt.grid(True)

plt.show()
```
