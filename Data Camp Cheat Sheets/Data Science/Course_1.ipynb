```python
#selecting rows that returns logic of true/false column

df.id > 500 # this returns records of rows (t/f) which are greater than 500
```

```python
# selecting rows based with logic returning the original value with all the rows and columns

df[df.id> 500] # this returns the original values instead of boolean of rows greater than 500
```

```python
# creating plots and adding details

plt.plot(df.id,df.id, label = 'test_label 1', color = 'DarkCyan', marker = 's') # for plotting number 1. 'label' shows the legend - this is added in every element of the plot (x,y axis combination)

# adding axes labels
plt.title('test title') # plot titled
plt.xlabel('id__2') # x axis title
plt.ylabel('id_1') # y axis title
plt.legend() #adds/shows legend based on the label of the plot above
plt.text(2.5,2,'test text') # adding arbitrary text to a plot (xcoord, ycoord, text )
plt.style.use('ggplot') # setting the overall style of the graph. Other styles (fivethirtyeight, seaborn)


plt.show()
```
