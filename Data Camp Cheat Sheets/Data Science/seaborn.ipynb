```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
student_data = pd.read_csv("D:\Data Science\Data Sources\Data Camp\student-alcohol-consumption.csv")

student_data.head()
student_data.info()
```

```python
sns.scatterplot(x = 'absences', y = 'G3', data = student_data, hue='location', hue_order=['Rural', 'Urban'])
```

```python
# dictionary for hue colors
palette_colors = {'Rural':'green', 'Urban':'blue'}

# plot
sns.countplot(x = 'school', data = student_data, hue='location', palette=palette_colors)

plt.show()
```

```python
# relational plots
    # allows you to create subplots in a single figure

sns.relplot(x = 'absences', y = 'G3', data = student_data, kind= 'scatter')
```

```python
# Modify the code to create one scatter plot for each level of the variable "study_time", arranged in columns.

# allows to create subplots per variable of the col
sns.relplot(x = 'absences', y = 'G3', data = student_data, kind= 'scatter', col = 'study_time')
```

```python
# Adapt your code to create one scatter plot for each level of a student's weekly study time, this time arranged in rows.
sns.relplot(x = 'absences', y = 'G3', data = student_data, kind= 'scatter', row = 'study_time')
```

```python
# Create a scatter plot of G1 vs. G3
sns.relplot(x = 'G1', y = 'G3', data = student_data, kind= 'scatter')
```

```python
# Create column subplots based on whether the student received support from the school ("schoolsup"), ordered so that "yes" comes before "no".
sns.relplot(x = 'G1', y = 'G3', data = student_data, kind= 'scatter', col  = 'schoolsup', col_order=('yes', 'no'))
```

```python
# Add row subplots based on whether the student received support from the family ("famsup"), ordered so that "yes" comes before "no". This will result in subplots based on two factors.
sns.relplot(x = 'G1', y = 'G3', data = student_data, kind= 'scatter', col  = 'schoolsup', col_order=('yes', 'no'), row= 'famsup', row_order= ('yes','no'))
```

```python
# categorical plots

# usage of catplot for categorical plots

survey_data = pd.read_csv("D:\Data Science\Data Sources\Data Camp\young-people-survey-responses.csv")

survey_data.head()
```

```python
sns.catplot(x ='Internet_usage', data = survey_data, kind = 'count')
```

```python
# horizontal 
sns.catplot(y ='Internet_usage', data = survey_data, kind = 'count')
```

```python
# subplot by age category younger than 21 and 21 and older

import numpy as np

# add new column for age category with np.where
survey_data['age_category'] = np.where(survey_data['Age'] >= 21, '21 and Older', 'Younger than 21')

sns.catplot(y ='Internet_usage', data = survey_data, kind = 'count', col = 'age_category')
```

```python
# Create a bar plot of interest in math, separated by gender

sns.catplot(x = 'Gender', y ='Mathematics', data  = survey_data, kind = 'bar')
```

```python
## Create bar plot of average final grade in each study category

sns.catplot(x = 'study_time', y = 'G3', data = student_data, kind = 'bar')
```

```python
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]

# reorder plot

sns.catplot(x = 'study_time', y = 'G3', data = student_data, kind = 'bar', order = category_order, ci = None)
```

```python
# box plots
    # distribution of quantitative data

# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]

# Create a box plot and set the order of the categories

sns.catplot(x = 'study_time', y = 'G3', data = student_data, kind = 'box', order = study_time_order)
```

```python
# subgroups and omit outliers

sns.catplot(x = 'internet', y = 'G3', data=student_data, kind = 'box', hue='location', sym = "")
```

```python
# Set the whiskers to 0.5 * IQR
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box", whis = 0.5)
plt.show()
plt.clf()

# Extend the whiskers to the 5th and 95th percentile
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[5,95])
plt.show()
plt.clf()
# Set the whiskers at the min and max values
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[0,100])
plt.show()
```

```python
# point plots

    # shows mean of quantitative variable

sns.catplot(x = 'famrel', y = 'absences', data = student_data, kind = 'point')

plt.show()
plt.clf()

# Add caps to the confidence interval
sns.catplot(x = 'famrel', y = 'absences', data = student_data, kind = 'point', capsize = 0.2)

plt.show()
plt.clf()

# Remove the lines joining the points
sns.catplot(x = 'famrel', y = 'absences', data = student_data, kind = 'point', capsize = 0.2, join = False)
```

```python
# Create a point plot that uses color to create subgroups
sns.catplot(x = 'romantic', y = 'absences', data = student_data, kind = 'point', hue = 'school')
# Show plot
plt.show()
plt.clf()

# turn off CI

sns.catplot(x = 'romantic', y = 'absences', data = student_data, kind = 'point', hue = 'school', ci= None)
plt.show()
plt.clf()

# Plot the median number of absences instead of the mean with numpy
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school",
            ci=None, estimator = np.median)
```

```python
# customizing sns plots

# Set the style to "whitegrid"
sns.set_style('whitegrid')
sns.set_palette('Purples')
sns.set_context('paper')

# Create a count plot of survey responses

sns.catplot(x="Internet_usage", 
            data = survey_data, 
            kind="count")

# Show plot
plt.show()
```

```python
# Set the style to "darkgrid"
sns.set_style('darkgrid')

# Set a custom color palette
sns.set_palette(["#39A7D0","#36ADA4"])

# Create the box plot of age distribution by gender
sns.catplot(x="Gender", y="Age", 
            data=survey_data, kind="box")

# Show plot
plt.show()
```

```python
# adding titles and axis labels

# determine first plot type - if FacetGrid or AxesSubplot through type()

    #axessubplot - contains only one variable per x and y axes (countplot, scatterplot, etc) - singleplot
    #facetgrid - contains one or more axes subplots (relplot, catplot) - can create subplots
```

```python
# identify sns plot type 

# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Identify plot type
type_of_g = type(g) # returns facetgrid (relplot)

# Print type
print(type_of_g)

# add title to facetgrid
g.fig.suptitle("Car Weight vs. Horsepower", y = 1.03) # y default: 1 - adjust to move title placement
```

```python
# add title to axesplot

h = sns.scatterplot(x = 'absences', y = 'G3', data = student_data, hue='location', hue_order=['Rural', 'Urban'])

type(h) # axes

h.set_title('Absences vs. Final Grade', y = 1.03)
```
