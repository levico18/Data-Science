```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("D:\Data Science\Data Sources\Data Sets\Diabetes\diabetes.csv")

print(df.head())
print(df.info())
```

```python
# test scatter

plt.scatter(df.Glucose,df.Insulin, color = 'Red', label = 'Insulin X Glucose', marker = 's', alpha = 0.1) # alpha shows transparency


plt.style.use('ggplot')
plt.title('Insulin and Glucose')
plt.legend()
plt.xlabel('Glucose')
plt.ylabel('Insulin')


plt.show()
```

```python
#creating bar chart

plt.bar(df.Age, df.Pregnancies)

plt.xlabel('Age')
plt.ylabel('Pregancies')
```

```python
#creating horizontal bar chart

plt.barh(df.Age, df.Pregnancies)
```

```python
# creating histogram

plt.hist(df.SkinThickness,bins=20, range = (20,50))

#setting to 20 bins with min and max range of 20 to 50
```

```python
#normalizing histograms

#unnormalized histogram

plt.hist(df.Insulin)
plt.hist(df.Glucose)
```

```python
#normalized histogram with 'density'
plt.hist(df.Insulin, density=True)
plt.hist(df.Glucose, density=True) 

# normalizing structures the dataset to the sum of 1 for an even proportio
```
